```python
%%manim ExampleScene -qm -v WARNING --disable_caching  -g


class ExampleScene(Scene):
    def construct(self):
        circle = Circle()
        square = Square()
        square.flip(RIGHT)
        square.rotate(-3 * TAU / 8)
        circle.set_fill(PINK, opacity=0.5)

        self.play(ShowCreation(square))
        self.play(Transform(square, circle))
```

```python
%%manim FameIdeaNo1  -p -qm -v WARNING --disable_caching   -s

class PreviewWindows(Group):

    def __init__(self):
        Group.__init__(self)
        height= 9/4
        width = 16/4
        s1 = Rectangle(height=height,width=width).set_color(BLUE)
        s2 = Rectangle(height=height,width=width).set_color(BLUE)
        s3 = Rectangle(height=height,width=width).set_color(BLUE)
        s4 = Rectangle(height=height,width=width).set_color(BLUE)
        s5 = Rectangle(height=height,width=width).set_color(BLUE)
        s6 = Rectangle(height=height,width=width).set_color(BLUE)

        a=VGroup(s1, s2, s3).set_x(0).arrange(buff=0.5).shift(UP)
        b=VGroup(s4, s5, s6).set_x(0).arrange(buff=0.5).shift(2 * DOWN)
        self.add(a,b)
        img= ImageMobject("media/images/first_point/ExampleScene10.png").move_to(s1).scale(0.42)
        img2= ImageMobject("media/images/first_point/ExampleScene20.png").move_to(s2).scale(0.42)
        img3= ImageMobject("media/images/first_point/ExampleScene30.png").move_to(s3).scale(0.42)
        img4= ImageMobject("media/images/first_point/ExampleScene40.png").move_to(s4).scale(0.42)
        img5= ImageMobject("media/images/first_point/ExampleScene45.png").move_to(s5).scale(0.42)
        img6= ImageMobject("media/images/first_point/ExampleScene50.png").move_to(s6).scale(0.42)

        self.add(img, img2, img3, img4, img5, img6)
        
class FameIdeaNo1(Scene):
    def construct(self):
        pw= PreviewWindows()
        self.add(pw)

```
